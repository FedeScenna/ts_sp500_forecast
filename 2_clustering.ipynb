```python
import pandas as pd
import numpy as np
from kshape.core import KShapeClusteringCPU 
from kshape.core_gpu import KShapeClusteringGPU 
```

```python
df = pd.read_csv("data/price_data.csv")
df.head()
```

```python
multivariate_ts_datasets = df.pivot(index='ticker', columns='date', values='close').values
```

```python
multivariate_ts_datasets
```

```python
num_clusters = 6

# CPU Model
ksc = KShapeClusteringCPU(num_clusters, max_iter=100, n_jobs=-1)
ksc.fit(multivariate_ts_datasets)

labels = ksc.labels_
cluster_centroids = ksc.centroids_
```
